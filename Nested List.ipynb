```python
if __name__ == '__main__':
    s=[]
    for _ in range(int(input())):
        name = input()
        score = float(input())
        l=[name,score]
        s.append(l)
    score=list()
    for i in s:
        for j in i:
            if isinstance(j, (int, float)):
                score.append(j)
    score = sorted(list(set(score)))
    final= score[1]
    for name,scr in sorted(s):
        if scr==final:
            print(name)
```
