```python
if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    d=student_marks[query_name]
    add=0
    for i in d:
        add=add+i
    avg=add/len(d)
    avg=round(avg,2)
    print(avg)
```
